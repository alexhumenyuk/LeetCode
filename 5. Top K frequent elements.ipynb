```python
class Solution:
    def topKFrequent(self, nums: list[int], k: int) -> list[int]:
        #Dict with nums as keys and their frequencies in the nums list as values
        freq_dict={}
        for num in nums:
            if num in freq_dict:
                freq_dict[num]+=1
            else:
                freq_dict[num]=1
        #print(freq_dict)
        nums_list=list(freq_dict.keys()) #keys as list
        freqs=list(freq_dict.values()) #values as list
        print("Numbers:")
        print(nums_list)
        print("Frequencies:")
        print(freqs)
        sorted_single_freqs=set(freqs) #convert the freqs list into a set to get unique freq values (eg.: [2,2,2,4,4,1] -> {1,2,4})
        sorted_single_freqs=sorted(sorted_single_freqs,reverse=True) #sort the freqs set in descending order, converting it back into a list: {1,2,4} -> [4,2,1]
        #print(sorted_single_freqs)
        #descend_freqs=[]
        corresp_nums=[]
        for single_freq in sorted_single_freqs: #iterate over the *sorted* list of unique frequencies to start with the highest frequency
            for i,freq in enumerate(freqs): #for each *unique* frequency, iterate over the list of frequencies. The frequency values in "freqs" have their corresponding number in "nums_list", meaning that freqs[i] corresponds to nums_list[i]
                if freq==single_freq:
                    if len(corresp_nums)<k: #keep only the first "k" numbers in the list
                        #descend_freqs.append(freq)
                        corresp_nums.append(nums_list[i]) #append the number corresponding to the current frequency to the numbers list
        #print("Descending frequencies:")
        #print(descend_freqs)
        print("Corresponding numbers:")
        print(corresp_nums)
        print("\n")
        return corresp_nums
sol=Solution()
nums = [1,1,1,2,2,3]
k = 2
print(sol.topKFrequent(nums,k))
nums = [1]
k = 1
print(sol.topKFrequent(nums,k))
nums=[3,0,1,0]
k=1
print(sol.topKFrequent(nums,k))
#[0]
nums=[3,2,3,1,2,4,5,5,6,7,7,8,2,3,1,1,1,10,11,5,6,2,4,7,8,5,6]
k=10
print(sol.topKFrequent(nums,k))
#[1,2,5,3,6,7,4,8,10,11]
```
